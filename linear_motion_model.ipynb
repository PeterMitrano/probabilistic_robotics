```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, u):
    """ The motion model
    x: [[x,y,theta], ...]
    u: [[linear_speed, angular_speed], ...]
    return: new state x_hat
    """
    x_hat = np.zeros(x.shape)
    x_hat[:,0] = x[:,0] + np.cos(x[:,2] + u[:,1]) * u[:,0]
    x_hat[:,1] = x[:,1] + np.sin(x[:,2] + u[:,1]) * u[:,0]
    x_hat[:,2] = x[:,2] + u[:,1]
    return x_hat
```

```python
M = 10000
x0_mu = [0, 0, np.deg2rad(0)]
x0_sigma = [0.01, 0.01, np.deg2rad(2)]
u_sigma = [0.05, np.deg2rad(15)]
u_mu = [1, np.deg2rad(5)]
x0 = np.random.randn(M,3) * x0_sigma + x0_mu
u = np.random.randn(M,2) * u_sigma + u_mu
x1 = f(x0, u)
plt.scatter(x0[:,0], x0[:,1], c='k', s=1)
plt.scatter(x0_mu[0], x0_mu[1], c='b', s=10)
plt.scatter(x1[:,0], x1[:,1], c='k', s=1)
plt.axis("equal")
```

```python
plt.title("Distributions of new state variables")
plt.hist(x1[:,2],bins=100, label='new theta', alpha=0.7)
plt.hist(x1[:,1],bins=100, label='new y')
plt.hist(x1[:,0],bins=100, label='new x')

plt.legend()
```
